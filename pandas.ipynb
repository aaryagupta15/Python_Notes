# <font color = "orange">Series & Data Frames</font>

```python
a = [10,9,56,7,3]
var = pd.Series(a)
print(var)
print(var[4])
```

```python
b = pd.Series([2,3,4], index = ['a','b','z'], dtype = float, name = "Aarya")
print(b)
```

```python
c = pd.Series([1,2,3,4,5,6,7])
d = pd.Series([1,2,3,4])

print(c+d)
```

```python
dic = {"name": ['Aarya','Ajita','Sanjeev','Nidhi'], "position": ['Daughter','Daughter','Father','Mother']}
print(dic)
print()
print(pd.Series(dic))
```

```python
e = [1,2,3,4,5,6,7]
dt = pd.DataFrame(e)
print(dt)
```

```python
di = {"name": ['Aarya','Ajita','Sanjeev','Nidhi'], "position": ['Daughter','Daughter','Father','Mother']}
# Length of all data should be same
dt1 = pd.DataFrame(di)
print(dt1)
```

```python
d = {"name": ['Aarya','Ajita','Sanjeev','Nidhi'], "position": ['Daughter','Daughter','Father','Mother']}
# Length of all data should be same
dt2 = pd.DataFrame(d,columns= ['name'], index = ['a','b','c','d'])
dt3 = pd.DataFrame(d,columns = ['position'])
print(dt2)
print()
print(dt3)
print()
print(d['name'][3])
```

```python
sen = {"s" : pd.Series([1,2,3,4,5]), "r":pd.Series([1,2,3,4,5])}
d2 = pd.DataFrame(sen)
print(d2)
```

```python
d2["New_Column"] = d2["s"]<=2
d2["New_Column2"] = d2["r"]<=3
print(d2)
```

```python
new_data = pd.DataFrame({"A": [1,2,3,4,5,6], "B": [1,2,3,4,5,6]})
print(new_data)
print()
new_data.insert(1,"insert_column",[4,5,6,7,7,8])
print(new_data)
```

```python
print(new_data.pop("insert_column"))
```

```python
del new_data["A"]
print(new_data)
```

# loc and iloc

**loc** : loc gets rows (and/or columns) with particular labels. <br>
**iloc** : iloc gets rows (and/or columns) at integer locations.

```python
df.loc[0,"Age"]="Python"
df
```

```python
df.loc[[0,1],["Name","Age"]]
```

```python
df.iloc[0, 1]
```
